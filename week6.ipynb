# 2.顯示x軸和y軸的說明標籤

```python
import pandas as pd
import matplotlib.pyplot as plt

# Colab 進行 matplotlib 繪圖時顯示繁體中文
# 下載台北思源黑體並命名為 taipei_sans_tc_beta.ttf，移至指定路徑
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download

import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
mpl.rc('font', family='Taipei Sans TC Beta')

products = {"分類": ["居家","居家","娛樂","娛樂","科技","科技"],
       "商店": ["家樂福A","大潤發A","家樂福B","全聯超A","大潤發B","家樂福C"],
       "價格": [11.42,23.50,19.99,15.95,55.75,111.55]}

df = pd.DataFrame(products, columns = ["分類", "商店", "價格"])

# 根據分類進行分組
grouped = df.groupby('分類')

# 繪製三種分類的折線圖
fig, ax = plt.subplots()
for name, group in grouped:
    ax.plot(group['商店'], group['價格'], marker='o', label=name)

ax.set_title("價格折線圖")  # 加上圖表標題
ax.set_xlabel("商店")  # 加上X軸標籤
ax.set_ylabel("價格")  # 加上Y軸標籤
plt.grid(True)  # 顯示圖表的格線
plt.xticks(rotation=45)  # 旋轉x軸刻度標籤
plt.legend(title='分類') # 加上圖例
plt.show()
```

# 3.請繪製水平長條圖

```python
import pandas as pd
import matplotlib.pyplot as plt

# Colab 進行 matplotlib 繪圖時顯示繁體中文
# 下載台北思源黑體並命名為 taipei_sans_tc_beta.ttf，移至指定路徑
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download

import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
mpl.rc('font', family='Taipei Sans TC Beta')

products = {"分類": ["居家","居家","娛樂","娛樂","科技","科技"],
       "商店": ["家樂福A","大潤發A","家樂福B","全聯超A","大潤發B","家樂福C"],
       "價格": [11.42,23.50,19.99,15.95,55.75,111.55]}

df = pd.DataFrame(products, columns = ["分類", "商店", "價格"])

# 根據分類進行分組
grouped = df.groupby('分類')

# 繪製水平長條圖
fig, ax = plt.subplots()
for name, group in grouped:
    ax.barh(group['商店'], group['價格'], label=name)

ax.set_title("價格變化水平長條圖")  # 加上圖表標題
ax.set_xlabel("價格")  # 加上X軸標籤
ax.set_ylabel("商店")  # 加上Y軸標籤
plt.grid(True)  # 顯示圖表的格線
plt.legend(title='分類') # 加上圖例
plt.show()
```

# 4.請繪製派圖

```python
import pandas as pd
import matplotlib.pyplot as plt

# Colab 進行 matplotlib 繪圖時顯示繁體中文
# 下載台北思源黑體並命名為 taipei_sans_tc_beta.ttf，移至指定路徑
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download

import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
mpl.rc('font', family='Taipei Sans TC Beta')

products = {
    "分類": ["居家", "居家", "娛樂", "娛樂", "科技", "科技"],
    "商店": ["家樂福A", "大潤發A", "家樂福B", "全聯超A", "大潤發B", "家樂福C"],
    "價格": [11.42, 23.50, 19.99, 15.95, 55.75, 111.55],
}

df = pd.DataFrame(products, columns=["分類", "商店", "價格"])

# 按照商店進行分組並計算每個商店的總價格
store_totals = df.groupby("商店")["價格"].sum()

# 繪製餅圖
plt.figure(figsize=(8, 8))  # 調整圖形大小
plt.pie(store_totals, labels=store_totals.index, autopct='%1.1f%%')
plt.title("商店價格分佈餅圖")
plt.show()
```
